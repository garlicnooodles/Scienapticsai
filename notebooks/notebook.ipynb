```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
cs_data=pd.read_csv('../data/Casestudy_dataset.csv')
```

```python
data_desc=pd.DataFrame()
data_desc['Metric']=['Min','Max','Mean','Median','1 percentile', '99 percentile']
for col,val in cs_data.items():
    min=val.min()
    max=val.max()
    mean=val.mean()
    median=val.median()
    q1=val.quantile(0.01)
    q99=val.quantile(0.99)
    a=[min,max,mean,median,q1,q99]
    data_desc[col]=a

pd.DataFrame(data_desc).style
```

```python
print(cs_data['V4'].value_counts().sort_index())
```

```python
corr=cs_data.corr()
corr_u=corr.unstack()

corr
```

```python


print("Variable with higher correlation (>= +-0.3)")
for col,val in corr_u.items():
    a,b=col
    if abs(val)>=0.3 and a!=b:
        print(f"{col}: {val:.4f}")
```

```python
bin_val=pd.interval_range(start=0, freq=500, end=20000, closed='right')
cs_data_binned_v6=pd.DataFrame(pd.cut(cs_data['V6'],bins=bin_val),dtype=str)
cs_data['V6_binned']=cs_data_binned_v6.replace({', ':'-'},regex=True)
cs_data['V6_binned']=cs_data['V6_binned'].str.strip('([])')
cs_data
```

```python
v6_bin_counts=cs_data['V6_binned'].value_counts()
col=v6_bin_counts.index
val=v6_bin_counts.values

fig,ax=plt.subplots(figsize=(40,20))
bars=ax.bar(col,val,width=0.6)
ax.bar_label(bars)
plt.show()

```

```python
loc=open('../data/Casestudy_json.json')
cs_data_json=json.load(loc)
```

```python
unpacked_json=[]
for quiz in cs_data_json:
    for sub in cs_data_json[quiz]: 
        for q_no in cs_data_json[quiz][sub]:
            i=1
            j=0
            for o in cs_data_json[quiz][sub][q_no]['options']:
                if o==cs_data_json[quiz][sub][q_no]['answer']:
                    j=i
                unpacked_json.append([quiz,sub,q_no,cs_data_json[quiz][sub][q_no]['question'],o,cs_data_json[quiz][sub][q_no]['answer'],j])
                i+=1
                

cs_data_json_upck=pd.DataFrame(unpacked_json, columns=['Title','Subject','Question_no','Question','Options','Answer','Answer_key'])

print(f"subjects and count of questions\n")
a=0
print(cs_data_json_upck['Question_no'].groupby(cs_data_json_upck['Subject']).agg('nunique'),'\n\n')

print(f"Answer keys\n")
print(cs_data_json_upck['Answer_key'].groupby(cs_data_json_upck['Question']).agg('max'),'\n\n')


```
